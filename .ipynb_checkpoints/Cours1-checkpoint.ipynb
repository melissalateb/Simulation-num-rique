```python
# Pour une fonction, on defini un interval [A...B], on donne des coordoneés, et un nombre de points a dessiner 
# 3 informations importantes
a = 1 # interval de début
b = 4 # interval de fin
n = 10 # nombre de points a dessiner
# xi = i *((b-a)/n) + a
L=[]
for i in range(n+1):
    L.append(a+i*(b-a)/n)
L
# X ??
# plot ??
plt.plot(L)
```

```python
a = 1 # interval de début
b = 4 # interval de fin
n = 10 # nombre de points a dessiner
# xi = i *((b-a)/n) + a
X = []
Y = []
def f(x): return x**2
for i in range(n+1):
    X.append(a+i*(b-a)/n)
    Y.append(f(a+i*(b-a)/n))
    
plt.plot(X,Y,"*-r")
```

```python
a = -2 # interval de début
b = 10 # interval de fin
n = 1000 # nombre de points a dessiner
# xi = i *((b-a)/n) + a
X = []
Y = []
def f(x): return np.sin(x**2)
for i in range(n+1):
    X.append(a+i*(b-a)/n)
    Y.append(f(a+i*(b-a)/n))
    
plt.plot(X,Y,"-r")
```

```python
import numpy as np

def f(x):
    if x>0: x**2
    else: return np.cos(x)
    
f(-2)
f(2)
```

```python
import matplotlib.pyplot as plt 
def f(x): return np.sin(x**2)
Vx = np.linspace(-2,10,10**7)
plt.plot(Vx, [f(x) for x in Vx], "-r")
```

```python
# Resolution d'équiation avec plusieurs inconnus
# 2x + 3y = 7
# 7x - 5y = 12
# assez facile a résoudre--> matrice associée
# x**2 - cos(y) = 7
# sqrt(x) + ln(y) = 9 
# equiation non linéaire 
# On demande a python de résoudre les équiation linéaire et non linéaire
```

```python
A = [[2,3], [7,-5]]
b = [7,12]
```

```python
from numpy.linalg import solve
```

```python
solve(A,b)
# array([2.29032258, 0.80645161]) --> tableau numpy
```

```python
# Multiplier une matrice par un vecteur en python (vérification du résultat)
A = np.array([[2,3], [7,-5]])
b = np.array([7,12])
sol = solve(A,b)
sol
A.dot(sol)-b # -->
```
