```python
x0, y0 = 1, 1
x1, y1 = 3, 4
x2, y2 = 5, 1

A = np.array([[1, x0, x0**2],
              [1, x1, x1**2],
              [1, x2, x2**2]])

b = np.array([y0, y1, y2])
res = np.linalg.solve(A, b)
a0, a1, a2 = res


x = np.linspace(x0, x2, 100)
y = a0 + a1 * x + a2 * x**2

plt.scatter([x0, x1, x2], [y0, y1, y2], color='red')
plt.plot(x, y)
print("a0 =", a0, ", a1 =", a1, ", a2 =", a2)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from numpy.linalg import solve

# def f(x) : return x**2 - 2 
# def fp(x): return 2*x
# n = 0
# def newton(x0):
# while n < 100:
# for i in range Nmax-1:
#     x= -f(x0)/f(x0) + x0
#     new_func = x0 - f(x)/fp(x)
#     return new_gen

def f(x):
    return np.tan(x) - x

def f_prime(x):
    return 1 / np.cos(x)**2 - 1

def newton_raphson(guess, tolerance=1e-6, max_iterations=100):
    x = guess
    for _ in range(max_iterations):
        x_next = x - f(x) / f_prime(x)
        if abs(x_next - x) < tolerance:
            return x_next
        x = x_next
    raise RuntimeError("La méthode de Newton n'a pas convergé.")

solution = newton_raphson(1)
print("La solution de tan(x) = x est approximativement:", solution)


x1=2
x2=3

Vx =np.linspace(-5,5,100)
fig, ax = plt.subplots()
ax.axhline(y=0, color='k')
ax.axvline(x=0, color='k')

ax.plot([x1,x2],[0,0], marker='*',color="red", ls='none', ms=7)
ax.plot(Vx, f(Vx), "-r")
plt.ylim(-3,3)
plt.show()
    
```

```python
import numpy as np
import matplotlib.pyplot as plt
from numpy.linalg import solve

def euler_explicite(y0, F, T, N):
    dt=T/N
    # Initialisation de la liste des valeurs de y
    y_values = [y0]
    for n in range(N):
        y_n = y_values[-1]
        F_n = F(y_n)
        y_next = y_n + dt * F_n
        y_values.append(y_next)

    return y_values

def F(y):
    return -y
y0 = 1
T  = 1
N  = 100

y_values = euler_explicite(y0, F, T, N)
print(y_values)

plt.plot(y_values,"xr")
```

```python
import numpy as np
import matplotlib.pyplot as plt

def F(y):
    return -y

def solve_differential_eq(y0, F, T, N, delta):
    dt = T / N
    y = np.zeros(N + 1)
    y[0] = y0

    for n in range(N):
        F_n = F(y[n])
        y[n + 1] = y[n] + dt * F_n
    return y

def exact_solution(y0, T, N):
    t = np.linspace(0, T, N+1)
    return y0 * np.exp(-t)

y0 = 1
T = 1
N = 50
delta = 0.1

y_values = solve_differential_eq(y0, F, T, N, delta)
exact_values = exact_solution(y0, T, N)

plt.plot(y_values, "xr", label='Approximation')
plt.plot(exact_values, "-b", label='Solution exacte')
# plt.plot(y_values, "xr")
# plt.plot(exact_values, "-b")

plt.legend()
```
